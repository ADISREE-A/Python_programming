```python
#466a
#n->total no. of travels
#m->no. of travels available using special ticket
#a-cost for normal ticket
#b cost for specil ticket
n,m,a,b=map(int,input().split())
if a*m<b:#check if cost of special ticket or normal ticket is cheap(here normal cheap)
    print(a*n)# all day travelled using special ticket only
else:
    print((n//m)*b+min(b,(n%m)*a))#special tickets cost+for remaining day calculate cost and which cost less is choosen
```

```python
#removing duplicate value
l=list(map(str,input().split()))
l1=[]
for ele in l:
    if ele not in l1:
        l1.append(ele)
print(l1)
```

```python
#no. of occurence odd and no duplicate
l=list(map(int,input().split()))
res=[]
for ele in l:
    if l.count(ele)%2!=0 and ele not in res:
        res.append(ele)
print(res)
```

```python
#even no. in desc and odd no. in asc ip:2 5 7 9 6 1 op:[6,2,1,5,7,9]
l=list(map(int,input().split()))
l.sort()
res=[]
for ele in l:
     if ele%2!=0:
        res.append(ele)
     else:
        res.insert(0,ele)
print(res)
```
